```python
import base64
from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()

kaggle_cfg = {}
for name in ['PRJ_RSA', 'NGROK_TOKEN_1', 'ID_RSA_PUB', 'SSH_PASS']:
    try:
        kaggle_cfg[name] = user_secrets.get_secret(name)
    except:
        pass

prj_rsa_val = kaggle_cfg.get("PRJ_RSA", "")
prj_rsa_val = base64.b64decode(prj_rsa_val).decode('ascii') if prj_rsa_val != "" else ""
ngrok_token_val = kaggle_cfg.get("NGROK_TOKEN_1", "")
id_rsa_pub      = kaggle_cfg.get("ID_RSA_PUB", "")
ssh_pass_val    = kaggle_cfg.get("SSH_PASS", "12345")
```
